```python
def download_video(video_id, url, out_path="videos/"):
    try:
        ydl_opts = {
            'outtmpl': os.path.join(out_path, f"{video_id}.%(ext)s"),
            'format': 'mp4',
            'quiet': True,
            'cookiefile': "cookies.txt",   # <<< pakai cookies login
        }
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            ydl.download([url])
        return True
    except Exception as e:
        print(f"Gagal download {video_id} -> {e}")
        return False
```

```python
import cv2
import os

# Folder tempat semua video berada
video_folder = "videos"
# Folder output untuk menyimpan frame
output_base = "frames"
os.makedirs(output_base, exist_ok=True)

# Ambil semua file video di folder
video_files = [f for f in os.listdir(video_folder) if f.endswith((".mp4", ".avi", ".mov"))]

for video_file in video_files:
    video_path = os.path.join(video_folder, video_file)
    # Buat folder per video
    video_name = os.path.splitext(video_file)[0]
    output_folder = os.path.join(output_base, video_name)
    os.makedirs(output_folder, exist_ok=True)

    cap = cv2.VideoCapture(video_path)
    frame_count = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        frame_filename = os.path.join(output_folder, f"frame_{frame_count:04d}.jpg")
        cv2.imwrite(frame_filename, frame)
        frame_count += 1

    cap.release()
    print(f"{video_file} selesai, total {frame_count} frame disimpan.")
```

```python
import cv2
import torch
import torchvision.models as models
import torchvision.transforms as transforms
from PIL import Image

# Load pretrained ResNet
resnet = models.resnet50(pretrained=True)
resnet = torch.nn.Sequential(*list(resnet.children())[:-1])  # ambil fitur sebelum fc
resnet.eval()
```

```python
# Transformasi untuk input ke ResNet
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

def extract_visual_features(video_file, frame_skip=30):
    cap = cv2.VideoCapture(video_file)
    features = []
    frame_count = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break
        if frame_count % frame_skip == 0:  # sampling frame
            img = Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
            x = transform(img).unsqueeze(0)
            with torch.no_grad():
                feat = resnet(x).squeeze().numpy()
            features.append(feat)
        frame_count += 1
    cap.release()
    return features
```

```python
import librosa
import numpy as np

def extract_audio_features(video_file, sr=22050, n_mfcc=13):
    y, sr = librosa.load(video_file, sr=sr)
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
    return np.mean(mfcc, axis=1)  # ambil rata-rata tiap koefisien
```

```python
def extract_frames(video_path, max_frames=30):
    frames = []
    cap = cv2.VideoCapture(video_path)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    step = max(1, frame_count // max_frames)

    i = 0
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        if i % step == 0:
            frame = cv2.resize(frame, (224, 224))  # Resize untuk CNN
            frames.append(frame)
        i += 1
    cap.release()
    return np.array(frames)

# Contoh ekstrak 1 video
frames = extract_frames("contoh_video.mp4")
print(frames.shape)  # (jumlah_frame, 224, 224, 3)
```
